```python
import time
class Nodo:
    def __init__(self, estado, hijo=None):
        self.estado = estado
        self.hijo = None
        self.padre = None
        self.accion = None
        self.acciones = None
        self.costo = None
        self.set_hijo(hijo)
    def set_estado(self, estado):
        self.estado = estado
    def get_estado(self):
        return self.estado
    def set_hijo(self, hijo):
        self.hijo = hijo
        if self.hijo is not None:
            for s in self.hijo:
                s.padre = self
    def get_hijo(self):
        return self.hijo
    def set_padre(self, padre):
        self.padre = padre
    def get_padre(self):
        return self.padre
    def set_accion(self, accion):
        self.accion = accion
    def get_accion(self):
        return self.accion
    def set_acciones(self, acciones):
        self.acciones = acciones
    def get_acciones(self):
        return self.acciones
    def set_costo(self, costo):
        self.costo = costo
    def get_costo(self):
        return self.costo
    def equal(self, Nodo):
        if self.get_estado() == Nodo.get_estado():
            return True
        else:
            return False
    def en_lista(self, lista_nodos):
        enlistado = False
        for n in lista_nodos:
            if self.equal(n):
                enlistado = True
        return enlistado
    def _str_(self):
        return str(self.get_estado())
```

```python
def busqueda_BPA_solucion(estado_inicial, solucion):

    resuelto = False
    nodos_visitados = []
    nodos_frontera = []
    nodo_raiz = Nodo(estado_inicial)
    nodos_frontera.append(nodo_raiz)
    while (not resuelto) and len(nodos_frontera) != 0:
        nodo_actual = nodos_frontera.pop(0)
        nodos_visitados.append(nodo_actual)
        if nodo_actual.get_estado() == solucion:
            resuelto = True
            return nodo_actual
        else:
            mis_hijos = []
            i = 0
            if nodo_actual.get_estado() == [0]*len(solucion):
                resuelto = True
                return nodo_actual
            while (i < len(nodo_actual.get_estado())-1):
                hijo = nodo_actual.get_estado().copy()
                auxi = hijo[i]
                hijo[i] = hijo[i+1]
                hijo[i+1] = auxi
                hijo_1 = Nodo(hijo)
                if not hijo_1.en_lista(nodos_visitados) and not hijo_1.en_lista(nodos_frontera):
                    nodos_frontera.append(hijo_1)
                mis_hijos.append(hijo_1)
                i = i + 1
            nodo_actual.set_hijo(mis_hijos)

def contar_digitos_posicion_final(estado_actual, posicion_final):
    conteo = 0
    for i in range(len(estado_actual)):
        if estado_actual[i] == posicion_final[i]:
            conteo += 1
    return conteo

def suma_distancias_posicion_final(estado_actual, posicion_final):
    suma_distancias = 0
    for i in range(len(estado_actual)):
        distancia = abs(estado_actual[i] - posicion_final[i])
        suma_distancias += distancia
    return suma_distancias

# heuristica añadida
def diferencia_absoluta_posicion_final(estado_actual, posicion_final):
    diferencia_absoluta = 0
    for i in range(len(estado_actual)):
        diferencia_absoluta += abs(estado_actual[i] - posicion_final[i])
    return diferencia_absoluta
```

```python
if __name__ == "__main__":

    estado_inicial = []
    n = int(input("Ingrese la cantidad de valores que desea agregar: "))
    for i in range(n):
        valor = int(input("Ingrese el valor {}: ".format(i+1)))
        estado_inicial.append(valor)
    solucion = []
    print(f"A que formacion de numeros quiere llegar: ")
    for i in range(n):
        valor = int(input("Ingrese el valor {}: ".format(i+1)))
        solucion.append(valor)
    inicio = time.time()

    # Calcular la heurística del conteo de dígitos en la posición final
    heuristica_conteo = contar_digitos_posicion_final(estado_inicial, solucion)
    print("Heurística del conteo de dígitos en la posición final:", heuristica_conteo)

    # Calcular la heurística de la suma de distancias a la posición final
    heuristica_distancias = suma_distancias_posicion_final(estado_inicial, solucion)
    print("Heurística de la suma de distancias a la posición final:", heuristica_distancias)

    # Calcular la heurística de la diferencia absoluta a la posición final
    heuristica_diferencia = diferencia_absoluta_posicion_final(estado_inicial, solucion)
    print("Heurística de la diferencia absoluta a la posición final:", heuristica_diferencia)

    nodo_solucion = busqueda_BPA_solucion(estado_inicial, solucion)
    resultado = []
    nodo_actual = nodo_solucion
    while nodo_actual.get_padre() is not None:
        resultado.append(nodo_actual.get_estado())
        nodo_actual = nodo_actual.get_padre()
    resultado.append(estado_inicial)
    resultado.reverse()
    print(resultado)
    fin = time.time()
    print(f"El tiempo de ejecución fue de {fin - inicio} segundos.")
```
